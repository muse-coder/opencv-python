```python
#图像处理
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,img_bi = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
ret,img_bi_inv = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY_INV)
ret,img_tr = cv2.threshold(img_gray,127,255,cv2.THRESH_TRUNC)
ret,img_zero = cv2.threshold(img_gray,127,255,cv2.THRESH_TOZERO)
ret,img_zero_inv = cv2.threshold(img_gray,127,255,cv2.THRESH_TOZERO_INV)
# plt.imshow(img)
titles = ['Original','Binary','Binary_INV','TRUNC','ZERO','ZERO_INV']
images = [img,img_bi,img_bi_inv,img_tr,img_zero,img_zero_inv]
plt.figure(figsize=(20,10))
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray'),plt.title(titles[i])
    plt.xticks([]),plt.yticks([])   # 不显示坐标轴
plt.show()
```

```python
# 图像处理 滤波
import cv2
def origin_demo(image):  # 原图
    cv2.imshow("origin", image)


def blur_demo(image):  # 均值滤波
    dst = cv2.blur(image, (3,3))
    cv2.imshow("avg_blur_demo", dst)


def median_blur_demo(image):  # 中值滤波
    dst = cv2.medianBlur(image,3)
    cv2.imshow("median_blur_demo", dst)


def bi_demo(image):  # 双边滤波
    dst = cv2.bilateralFilter(image, 0, 100, 5)
    cv2.imshow("bi_demo", dst)


def shift_demo(image):  # 均值迁移
    dst = cv2.pyrMeanShiftFiltering(image, 10, 50)
    cv2.imshow("shift_demo", dst)

def gaussian_demo(image):
    dst=cv2.GaussianBlur(image,(3,3),1)
    cv2.imshow("gaussian_demo", dst)

img = cv2.imread("face.jpg", 1)
blur_demo(img)
origin_demo(img)
median_blur_demo(img)
bi_demo(img)
shift_demo(img)
gaussian_demo(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
